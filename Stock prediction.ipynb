```python
df=pd.read_csv('FSE-PUM_X.csv')
```

```python
df.drop(axis=1,columns=['Last Price of the Day','Daily Traded Units','Daily Turnover','Change'],inplace=True)
```

```python
df.dropna(axis=0,inplace=True)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = np.array(df.loc[:,['High','Close']]) #data
y = np.array(df.loc[:,'Turnover']) #outcome
```

```python
# dividing X, y into train and test data 
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.25,random_state = 10) 
```

```python
# training a linear regression classifier 
from sklearn.linear_model import LinearRegression 
lreg = LinearRegression()
lreg.fit(X_train, y_train)
```

```python
# accuracy on X_test 
accuracy = lreg.score(X_test, y_test) 
print(accuracy) #accuracy is poor
```

```python
y_pred=lreg.predict(X_test)
```

```python
#plotting
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.scatter(X_test[:,0],y_test,color='orange')
ax.plot(X_test[:,0],y_pred,color='blue')
iplot_mpl(fig,resize=True)
```
